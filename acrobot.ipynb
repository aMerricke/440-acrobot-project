```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt

env = gym.make('Acrobot-v1', render_mode=None)
observation, info = env.reset(seed=123)

episode_over = False
total_reward = 0
cum_reward = []

observations = []
while not episode_over:
    action = env.action_space.sample()
    observation, reward, terminated, truncated, info = env.step(action)
    
    total_reward += reward
    cum_reward.append(total_reward)
    episode_over = terminated or truncated

    observations.append(observation)
    
print(f"Episode finished. Total reward: {total_reward}")
env.close()
```

```python
steps = range(len(observations))


# plt.plot(steps, observations)

costh1 = [observation[0] for observation in observations]
sinth1 = [observation[1] for observation in observations]
costh2 = [observation[2] for observation in observations]
sinth2 = [observation[3] for observation in observations]
omega1 = [observation[4] for observation in observations]
omega2 = [observation[5] for observation in observations]
```

```python
plt.plot(steps, sinth1)
plt.title("sinth1 vs. step")
plt.xlabel("step")
plt.ylabel("sinth1 [rad]")
plt.savefig("sin1_v_step.png")
```

```python
plt.plot(steps, omega1)
plt.title("omega1 vs. step")
plt.xlabel("step")
plt.ylabel("omega1 [rad/s]")
plt.savefig("omega1_v_step.png")
```

```python
plt.plot(steps, cum_reward)
plt.title("reward vs step")
plt.xlabel("step")
plt.ylabel("cumulative reward")
plt.savefig("reward_v_step.png")
```
